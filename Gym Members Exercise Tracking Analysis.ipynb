#Workout Type Preference By Gender.

```python
workout_gender_counts = df.groupby(['Workout_Type', 'Gender'])['Gender'].count().reset_index(name='Count')
```

```python
fig = px.bar(workout_gender_counts,
             x='Workout_Type',
             y='Count',
             color='Gender',
             title='Workout Type Preference by Gender')
fig.show()
```

#Average Weight (kg) And Height (m) By Gender (Ages 18-35)

```python
filtered_df = df[(df['Age'] >= 18) & (df['Age'] <= 35)]
```

```python
gender_stats = filtered_df.groupby('Gender')[['Height (m)', 'Weight (kg)']].mean()
```

```python
fig = go.Figure(data=[
    go.Bar(name='Weight (kg)', x=gender_stats.index, y=gender_stats['Weight (kg)']),
    go.Bar(name='Height (m)', x=gender_stats.index, y=gender_stats['Height (m)'])
])

fig.update_layout(barmode='group', title_text='Average Weight (kg) and Height (m) by Gender (Ages 18-35)')
fig.show()
```

#Average Weight (kg) And Height (m) By Gender (Ages 36-59)

```python
filtered_df = df[(df['Age'] >= 36) & (df['Age'] <= 59)]
```

```python
gender_stats = filtered_df.groupby('Gender')[['Height (m)', 'Weight (kg)']].mean()
```

```python
fig = go.Figure(data=[
    go.Bar(name='Weight (kg)', x=gender_stats.index, y=gender_stats['Weight (kg)']),
    go.Bar(name='Height (m)', x=gender_stats.index, y=gender_stats['Height (m)'])
])

fig.update_layout(barmode='group', title_text='Average Weight (kg) and Height (m) by Gender (Ages 36-59)')
fig.show()
```

#BPM Statistics By Gender (Ages 18-35)

```python
filtered_df = df[(df['Age'] >= 18) & (df['Age'] <= 35)]
```

```python
gender_bpm_stats = filtered_df.groupby('Gender')[['Max_BPM', 'Avg_BPM', 'Resting_BPM']].agg(['max', 'mean'])
```

```python
fig = go.Figure()

for bpm_type in ['Max_BPM', 'Avg_BPM', 'Resting_BPM']:
    fig.add_trace(go.Scatter(x=gender_bpm_stats.index,
                             y=gender_bpm_stats[bpm_type]['mean'],
                             mode='lines+markers',
                             name=f'{bpm_type} (Avg)'))
    fig.add_trace(go.Scatter(x=gender_bpm_stats.index,
                             y=gender_bpm_stats[bpm_type]['max'],
                             mode='lines+markers',
                             name=f'{bpm_type} (Max)'))

fig.update_layout(title='BPM Statistics by Gender (Ages 18-35)',
                  xaxis_title='Gender',
                  yaxis_title='BPM')

fig.show()
```

#BPM Statistics By Gender (Ages 36-59)

```python
filtered_df = df[(df['Age'] >= 36) & (df['Age'] <= 59)]
```

```python
gender_bpm_stats = filtered_df.groupby('Gender')[['Max_BPM', 'Avg_BPM', 'Resting_BPM']].agg(['max', 'mean'])
```

```python
fig = go.Figure()

for bpm_type in ['Max_BPM', 'Avg_BPM', 'Resting_BPM']:
    fig.add_trace(go.Scatter(x=gender_bpm_stats.index,
                             y=gender_bpm_stats[bpm_type]['mean'],
                             mode='lines+markers',
                             name=f'{bpm_type} (Avg)'))
    fig.add_trace(go.Scatter(x=gender_bpm_stats.index,
                             y=gender_bpm_stats[bpm_type]['max'],
                             mode='lines+markers',
                             name=f'{bpm_type} (Max)'))

fig.update_layout(title='BPM Statistics by Gender (Ages 36-59)',
                  xaxis_title='Gender',
                  yaxis_title='BPM')

fig.show()
```

#Total Water Intake By Gender And Water Intake (Liters).

```python
workout_gender_water = df.groupby(['Water_Intake (liters)', 'Gender'])['Water_Intake (liters)'].sum().reset_index(name='Total_Water_Intake')
```

```python
fig = px.bar(workout_gender_water,
             x='Water_Intake (liters)',
             y='Total_Water_Intake',
             color='Gender',
             title='Total Water Intake by Gender and Water Intake (liters)')
fig.show()
```

#Fat Percentage Count By Gender And Experience Level.

```python
fat_percentage_counts = df.groupby(['Gender', 'Experience_Category', 'Fat_Percentage'])['Fat_Percentage'].count().reset_index(name='Count')
```

```python
fig = px.line(fat_percentage_counts,
              x='Fat_Percentage',
              y='Count',
              color='Gender',
              line_group='Experience_Category',
              title='Fat Percentage Count by Gender and Experience Level',
              labels={'Fat_Percentage': 'Fat Percentage (%)', 'Count': 'Number of Members'})
fig.show()
```
